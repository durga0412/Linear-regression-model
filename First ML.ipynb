```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\DELL\Downloads\house_data.csv")
df.head(10)
```

```python
df.describe()
```

```python
plt.scatter(df["sqft_living"], df["price"])
```

```python
y = df["price"]
x = df.drop(["price", "id", "date"], axis = 1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test) #for future data we can call model.predict() and get the house price of a given input data
#here efficiency is 71.56%
```
